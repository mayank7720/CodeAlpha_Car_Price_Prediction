# Exploratory Data Analysis - Car Price Prediction
This notebook walks through the data collection, cleaning, visualization, and distribution analysis of used car listings.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the raw dataset
df = pd.read_csv('../dataset/raw/car_data.csv')
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")
df.head()
```

### 1. Data Integrity and Cleaning

```python
# Check data types & missing values
print(df.info())
print("Missing values per column:\n", df.isnull().sum())

# Drop duplicate entries
initial_len = len(df)
df = df.drop_duplicates()
print(f"Removed {initial_len - len(df)} duplicate rows. New dataset size: {len(df)}")
```

### 2. Feature Distribution Visualizations using Matplotlib

```python
# Plot Selling Price Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['Selling_Price'], kde=True, color='red', bins=30)
plt.title('Distribution of Resale Prices (Lakhs)')
plt.xlabel('Selling Price (Lakhs)')
plt.ylabel('Count')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Scatter plot with trendline (Present Price vs Selling Price)
plt.figure(figsize=(10, 6))
plt.scatter(df['Present_Price'], df['Selling_Price'], color='blue', alpha=0.6, label='Car Listings')
m, c = np.polyfit(df['Present_Price'], df['Selling_Price'], 1)
plt.plot(df['Present_Price'], m * df['Present_Price'] + c, color='red', linewidth=2, label=f'Trendline (Slope: {m:.2f})')
plt.title('Resale Value vs Initial Showroom Price')
plt.xlabel('Showroom Price (Present Price in Lakhs)')
plt.ylabel('Resale Price (Selling Price in Lakhs)')
plt.legend()
plt.grid(True, alpha=0.2)
plt.show()
```

### 3. Correlation Analysis

```python
# Numerical Feature Correlation Matrix Heatmap
plt.figure(figsize=(8, 6))
num_cols = ['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven', 'Owner']
corr_matrix = df[num_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Numerical Variables Pearson Correlation Heatmap')
plt.show()
```
